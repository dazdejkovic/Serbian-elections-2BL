# Serbian presidental elections and Benford law

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import benford as bf
```

## Presidental elections 2002

```python
sd = pd.read_csv('Podaci/P_2002_6_1a.csv')
sd=sd.astype(int)
sd.head(2)
```

Choosing 3 top results in first round

```python
top = ['D', 'E', 'K' ]
```

```python
for K in top:
    benf = bf.Benford((sd, K), decimals=0) # Initialisation of Benford object
    benf.SD.report()
    print(K, ': jhattest= ',abs((benf.SD['Found']*np.arange(10)).sum()-4.187))  # Expected value of second digit    
```

# Presidental elections 2003

```python
sd = pd.read_csv('Podaci/P_2003_6_1.csv')
sd = sd[pd.to_numeric(sd['A'], errors='coerce').notnull()]   # Cleaning data of non numeric rows
sd=sd.astype(int)
sd.head(1)
```

```python
top = ['D', 'C', 'B' ]
for K in top:
    benf = bf.Benford((sd, K), decimals=0, limit_N=1800) # Initialisation of Benford object
    benf.SD.report()
    print(K, ': jhattest= ',abs((benf.SD['Found']*np.arange(10)).sum()-4.187))  # Expected value of second digit
```

## Presidental elections 2004
