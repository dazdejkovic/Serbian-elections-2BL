# Serbian presidental elections and Benford law

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import benford as bf
import scipy.stats as stats
```

## Calculation of probability's distribution for Benford law for secend significant digit

```python
r=np.array([ sum([np.log10(1+1/(10*k+j)) for k in range(1,10)]) for j in range(10)])
r
```

```python
j_hat=(r*range(10)).sum()  # Mean of second significant digit
j_hat
```

## Presidental elections 2002

```python
sd = pd.read_csv('Podaci/P_2002_6_1a.csv')
sd=sd.astype(int)
sd.head(2)
```

Choosing 3 top results in first round

```python
top = ['D', 'E', 'K' ]
```

```python
for K in top:
    print(30*K)
    benf = bf.Benford((sd, K), decimals=0) # Initialisation of Benford object
    benf.SD.report()
    print()
    print(stats.ttest_1samp(sd[K], popmean=j_hat))    
    print()
```
